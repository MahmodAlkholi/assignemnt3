## Single Indexing

**Get the thered row with fourth column**

```python
df.loc[3 , "age"]
```

# Muultible indexing

```python
rows = np.arange( 1 , 5 )
col = "age"

df.loc[ rows, col]
```

```python
rows = np.arange( 1 , 5 )
col = 4

df.iloc[ rows, col]
```
